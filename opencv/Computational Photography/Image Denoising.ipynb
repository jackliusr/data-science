```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as ssim
import urllib
```

```python
req = urllib.request.urlopen('https://raw.githubusercontent.com/jackliusr/data-science/main/data/die.png')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img  = cv.imdecode(arr, cv.IMREAD_COLOR ) # 'Load it as grayscale'
# gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
dst = cv.fastNlMeansDenoisingColored(img,None,10,10,7,21)
plt.figure(figsize=(15,15))
plt.subplot(121),plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.subplot(122),plt.imshow(cv.cvtColor(dst,cv.COLOR_BGR2RGB))
plt.show()
```
