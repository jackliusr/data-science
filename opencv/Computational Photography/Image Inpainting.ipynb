```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as ssim
import urllib
```

```python
req = urllib.request.urlopen('https://raw.githubusercontent.com/makelove/OpenCV-Python-Tutorial/master/data/messi_2.jpg')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img  = cv.imdecode(arr, cv.IMREAD_COLOR ) # 'Load it as grayscale'
# gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)

req = urllib.request.urlopen('https://raw.githubusercontent.com/makelove/OpenCV-Python-Tutorial/master/data/mask2.png')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
mask   = cv.imdecode(arr, cv.IMREAD_GRAYSCALE )
```

```python
dst = cv.inpaint(img,mask,3,cv.INPAINT_TELEA)

fig, axs = plt.subplots(2, 2, figsize=(15,15))
axs[0, 0].imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
axs[0, 0].set_title('img')
axs[0, 1].imshow(mask, cmap="gray")
axs[0, 1].set_title('mask')
axs[1, 0].imshow(cv.cvtColor(dst,cv.COLOR_BGR2RGB))
axs[1, 0].set_title('Dest')
axs[1, 1].imshow(cv.cvtColor(dst,cv.COLOR_BGR2RGB))
axs[1, 1].set_title('Dest')

plt.show()
```
