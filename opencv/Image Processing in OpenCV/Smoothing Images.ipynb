```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
```

```python
# 2D Convolution ( Image Filtering )

img = imutils.url_to_image('https://raw.githubusercontent.com/opencv/opencv/master/samples/data/opencv-logo.png')
kernel = np.ones((5,5),np.float32)/25
dst = cv.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
from skimage.metrics import structural_similarity as ssim

(score, diff) = ssim(img, dst, full=True, multichannel=True)
print("Image similarity:", score)

diff = (diff * 255).astype("uint8")
plt.imshow(diff)
```

```python
# Image Blurring (Image Smoothing)

img = imutils.url_to_image('https://raw.githubusercontent.com/opencv/opencv/master/samples/data/opencv-logo-white.png')
blur = cv.blur(img,(5,5))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()

plt.figure()
(score, diff) = ssim(img, blur, full=True, multichannel=True)
print("Image similarity:", score)
diff = (diff * 255).astype("uint8")
plt.imshow(diff)
```

```python
# 2. Gaussian Blurring
blur = cv.GaussianBlur(img,(5,5),0)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()

plt.figure()
(score, diff) = ssim(img, blur, full=True, multichannel=True)
print("Image similarity:", score)

diff = (diff * 255).astype("uint8")
plt.imshow(diff)
```

```python
# Median Blurring
blur = cv.medianBlur(img,5)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()

plt.figure()
(score, diff) = ssim(img, blur, full=True, multichannel=True)
print("Image similarity:", score)

diff = (diff * 255).astype("uint8")
plt.imshow(diff)
```

```python
# Bilateral Filtering
blur = cv.bilateralFilter(img,9,75,75)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()

plt.figure()
(score, diff) = ssim(img, blur, full=True, multichannel=True)
print("Image similarity:", score)

diff = (diff * 255).astype("uint8")
plt.imshow(diff)
```
