```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
from skimage.metrics import structural_similarity as ssim
```

```python
img = imutils.url_to_image("https://raw.githubusercontent.com/advcha/OpenCV-Python/master/images/bolt.png")
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY);
ret,thresh = cv.threshold(img,127,255,0)
contours,hierarchy = cv.findContours(thresh,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)

(x,y),radius = cv.minEnclosingCircle(contours[0])
center = (int(x),int(y))
radius = int(radius)
print(center)
print(radius)

img_circle = img.copy()


img_circle= cv.circle(img_circle,center,radius,(255,255,0),2)

plt.figure()
plt.imshow(img)
plt.figure()
plt.imshow(img_circle)

(score, diff) = ssim(img, img_circle, full=True)
print("Image similarity:", score)


diff = (diff * 255).astype("uint8")
plt.figure()
plt.imshow(diff)
plt.show()
```
