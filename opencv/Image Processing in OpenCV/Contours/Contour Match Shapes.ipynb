```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
from skimage.metrics import structural_similarity as ssim
```

```python
# https://raw.githubusercontent.com/advcha/OpenCV-Python/master/images/approx.jpg

img1  = imutils.url_to_image("https://raw.githubusercontent.com/advcha/OpenCV-Python/master/images/star.jpg")
img2  = imutils.url_to_image("https://raw.githubusercontent.com/advcha/OpenCV-Python/master/images/star2.jpg")
plt.imshow(img1)
plt.figure()
plt.imshow(img2)

img1 =cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
img2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)

ret, thresh = cv.threshold(img1, 127, 255,0)
ret, thresh2 = cv.threshold(img2, 127, 255,0)
contours,hierarchy = cv.findContours(thresh,2,1)
cnt1 = contours[0]
contours,hierarchy = cv.findContours(thresh2,2,1)
cnt2 = contours[0]
ret = cv.matchShapes(cnt1,cnt2,1,0.0)
print( ret )

plt.figure()
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
for ax, interp,img in zip(axs, ['Star1', 'Star2'],[img1,img2]):
    ax.imshow(img)
    ax.set_title(interp.capitalize())
    ax.grid(False)
plt.show()
```
