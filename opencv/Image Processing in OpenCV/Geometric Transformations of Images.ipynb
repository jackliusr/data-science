```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
```

```python
img = imutils.url_to_image('https://raw.githubusercontent.com/opencv/opencv/master/samples/data/messi5.jpg')

print(img.shape)

res = cv.resize(img,None,fx=2, fy=2, interpolation = cv.INTER_CUBIC)
plt.imshow(res)
plt.show()

res2= cv.resize(img,(548*2, 342*2), interpolation = cv.INTER_CUBIC)
plt.figure()
plt.imshow(res2)
plt.show()
```

```python
# Translation
rows,cols, _ = img.shape
M = np.float32([[1,0,100],[0,1,50]])
dst = cv.warpAffine(img,M,(cols,rows))
plt.imshow(dst)
plt.show()
```

```python
# Rotation
rows,cols,_ = img.shape
# cols-1 and rows-1 are the coordinate limits.
M = cv.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),90,1)
dst = cv.warpAffine(img,M,(cols,rows))
plt.imshow(dst)
plt.show()
```

```python
# Affine Transformation
img = imutils.url_to_image('https://raw.githubusercontent.com/opencv/opencv/master/samples/data/sudoku.png')
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv.getAffineTransform(pts1,pts2)
dst = cv.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
# Perspective Transformation
img = imutils.url_to_image('https://raw.githubusercontent.com/opencv/opencv/master/samples/data/sudoku.png')
rows,cols,ch = img.shape
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv.getPerspectiveTransform(pts1,pts2)
dst = cv.warpPerspective(img,M,(300,300))
plt.subplot(121),plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)),plt.title('Input')
plt.subplot(122),plt.imshow(cv.cvtColor(dst,cv.COLOR_BGR2RGB)),plt.title('Output')
plt.show()
```
