```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
from skimage.metrics import structural_similarity as ssim
```

```python
img = imutils.url_to_image('https://raw.githubusercontent.com/opencv/opencv/master/samples/data/messi5.jpg')
```

```python
plt.imshow(img)
lower_reso = cv.pyrDown(img)
plt.figure()
plt.imshow(lower_reso)
lower_reso2 = cv.pyrDown(lower_reso)
plt.figure()
plt.imshow(lower_reso2)
higher_reso2 = cv.pyrUp(lower_reso2)
plt.figure()
plt.imshow(higher_reso2)
```
