```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
from skimage.metrics import structural_similarity as ssim
```

```python
# Erosion
img = imutils.url_to_image('https://docs.opencv.org/master/j.png')
plt.imshow(img)
kernel = np.ones((5,5),np.uint8)
erosion = cv.erode(img,kernel,iterations = 1)
plt.figure()
plt.imshow(erosion)
plt.show()
```

```python
# Dilation
dilation  = cv.dilate(img,kernel,iterations = 1)
plt.figure()
plt.imshow(dilation )
plt.show()
```

```python
# Opening
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
plt.figure()
plt.imshow(opening)
plt.show()

plt.figure()
(score, diff) = ssim(img, opening, full=True, multichannel=True)
print("Image similarity:", score)
diff = (diff * 255).astype("uint8")
plt.imshow(diff)
```

```python
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)
gradient = cv.morphologyEx(img, cv.MORPH_GRADIENT, kernel)
tophat = cv.morphologyEx(img, cv.MORPH_TOPHAT, kernel)
blackhat = cv.morphologyEx(img, cv.MORPH_BLACKHAT, kernel)


fig=plt.figure()

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20,20))
axes[0,0].axis("off")
axes[0,0].imshow(closing,  aspect="auto")
axes[0,1].axis("off")
axes[0,1].imshow(gradient,  aspect="auto")
axes[1,0].axis("off")
axes[1,0].imshow(tophat,  aspect="auto")
axes[1,1].axis("off")
axes[1,1].imshow(blackhat,  aspect="auto")
plt.subplots_adjust(wspace=.05, hspace=.05)
plt.show()
```
