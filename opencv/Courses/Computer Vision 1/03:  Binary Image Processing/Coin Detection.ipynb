```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as ssim
import urllib
```

```python
req = urllib.request.urlopen('https://raw.githubusercontent.com/spmallick/learnopencv/master/BlobDetector/blob.jpg')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv.imdecode(arr,cv.IMREAD_GRAYSCALE) # 'Load it as grayscale'
gray =img.copy()
plt.imshow(gray, cmap="gray")
plt.show()
```

```python
detector = cv.SimpleBlobDetector_create()
```

```python
keypoints = detector.detect(gray)
```

```python
im_with_keypoints = cv.drawKeypoints(gray, keypoints, np.array([]), (0,0,255), cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.imshow(im_with_keypoints, cmap="gray")
plt.show()
```
